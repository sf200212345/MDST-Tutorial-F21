<h2> Feature Selection </h2>

After exploring the data, you now have to choose features that you would use to predict the performance of the students on a chosen test (your chosen predictor). By the way, you can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features

```python
statesEd["TOTAL_REVENUE_PER_STUDENT"] = statesEd["TOTAL_REVENUE"] * (statesEd["GRADES_8_G"] / statesEd["GRADES_ALL_G"])
statesEd["I+S_EXPENDITURE_PER_STUDENT"] = (statesEd["SUPPORT_SERVICES_EXPENDITURE"] + statesEd["INSTRUCTION_EXPENDITURE"]) * (statesEd["GRADES_8_G"] / statesEd["GRADES_ALL_G"])
```

Final feature list: **<Total Revenue Per Student, Instruction + Support Services Expenditure Per Student\>**

Feature selection justification: **<First, when I refer to student, since the data concerns 8th grade students, I'm referring specifically to 8th grade students. Thus, my calculation of total revenue per student takes into account the proportion of the 8th grade student population relative to the overall student population, as there can be fluctuations in the population of 8th grade students that affects how much, on average, each student will get in terms of revenue and expenditure. Total revenue per student is meant to measure how much revenue the state receives per student with the assumption that the size of this revenue will affect the educational experience of the student. Similarly, I+S expenditure per student is meant to show how much is actually spent on important student services in math, as instruction and support services will probably affect students more than capital outlay expenditures specifically in math. \>**

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
#considering 4 variables based on the graphs from before, this time including year since the way money is spent changes over time
X = statesEd[["TOTAL_REVENUE_PER_STUDENT","GRADES_8_G","I+S_EXPENDITURE_PER_STUDENT", "YEAR"]]
#classification task to see if these characteristics are enough to determine if students can get to a national average score
y = statesEd.loc[X.index]["AVG_MATH_8_SCORE"]> statesEd["AVG_MATH_8_SCORE"].mean()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.3, random_state=0)
```

<h2> Prediction </h2>

ML Models Resource: https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a

Chosen ML task: **<REGRESSION/CLASSIFICATION>**

```python
# import your sklearn class here
from sklearn.neighbors import KNeighborsClassifier
#since we want to find if other students with similar profiles will get similar scores, we use KNN
```

```python
# create your model here
model = KNeighborsClassifier(n_neighbors=10) #more neighbor comparisons hopefully gives us more accurate results
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# FOR CLASSIFICATION ONLY:
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(model, X_test, y_test,
                         cmap=plt.cm.Blues)
```

<h2> Summary </h2>

```python
from sklearn.metrics import accuracy_score, precision_score
```

```python
print("ACCURACY: ", accuracy_score(y_test, model.predict(X_test)))
print("PRECISION: ", precision_score(y_test, model.predict(X_test)))
```

**<In this Machine Learning exercise, I used a KNN classifier to see how well we could predict whether an 8th grader would be able to score above the average (as calculated by taking the average across all the scores in the AVG_MATH_8_SCORE column) when taking into consideration the year, size of the 8th grade student body, total revenue per 8th grader and total expenditure on instruction and support services per 8th grader. Below are the justifications for these parameters:

Year: As noted above, the year can impact how much total revenue is accepted by a state based on the state's needs and its population fluctuations, as well as how efficiently money is spent on instructional and support services.

Size of 8th grade student body: The idea here is that a larger 8th grade student body will average out the math scores more so than smaller student bodies, so people from larger 8th grade student bodies will more likely score closer to the mean. Additionally, as seen from the two graphs, schools with larger 8th grade student bodies generally will get more revenue and spend more on each student.

Total revenue per 8th grader and total expenditure on instruction and support services per 8th grader: In addition to the justification above, as we can see in the graph, higher levels of total revenue per student and total expenditure per student shows a tighter concentration of data points around the mean when compared to the data points at lower levels of these factors. Thus, it would appear that spending more and earning more revenue per student creates a general trend of better academic success.

From the actual machine learning part, we can see from the confusion matrix that the classifier used was very good at determining whether a student would achieve scores higher than the average based on the given factors, classifying 61 / 72 (around 85%) of data points that were above the average to be above the average. However, the classifier was terrible at determining which data points would be under the average, with a classification success rate at 29 / 60 (around 48%). This resulted in the overall accuracy of the classifier to be around 0.68, and its precision is similarly around 0.66, showing that while these factors do have an impact on whether students will score above the average, it is not an incredibly significant impact. This result is consistent with the spread of the graphs, as while there is a general upward trend of these factors leading to increased average math scores, the data points are all still very spread out to only be able to suggest general correlation.
